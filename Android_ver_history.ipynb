```python
from urllib.request import urlopen
```

```python
url="https://en.wikipedia.org/wiki/Android_version_history"
```

```python
#get request
data=urlopen(url)
type(data)
```

```python
html=data.read()
```

```python
html
```

```python
data.close()
```

```python
from bs4 import BeautifulSoup as soup
#parse html
```

```python
android_soup=soup(html,'html.parser')
android_soup
```

```python
android_soup.h1
```

```python
android_soup.find_all('h1',{})
```

```python
#filter relevent parts
tables = android_soup.find_all("table",{'class':'wikitable'})
len(tables)
```

extracting useful info 
-> remove undesired tags
->extract table header and data


```python
android_table=tables[0]
headers = android_table.find_all('th')
headers[0].text[:-1]
```

```python
columns=[col.text[:-1] for col in headers]
columns
```

```python
rows=android_table.find_all('tr')
print(rows[0])
```

```python

table_rows=[]
for row in rows:
    current_row=[]
    row_data=row.find_all('td')
    for idx,data in enumerate(row_data):
        current_row.append(data.text[:-1])
    table_rows.append(current_row)
```

```python
print(table_rows)
```

Writing and Reading the csv file

```python
filename = 'android_version_history.csv'
with open(filename, 'w', encoding='utf-8') as f:
    # Writing header
    header_string = ','.join(columns)
    header_string += '\n'
    f.write(header_string)

    for row in table_rows[:-1]:
        row_string = ""
        for w in row:
            w=w.replace(',','')
            row_string+=w+','
        row_string += '\n'
        f.write(row_string)
```
